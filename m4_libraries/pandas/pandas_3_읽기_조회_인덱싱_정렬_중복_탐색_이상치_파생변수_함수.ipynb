```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
df[['sibsp','parch']].head()
```

```python
def add(x,y):
  return x + y
df['family'] = df.apply(add(df['sibsp'],df['parch']))
df.head()
# df[['sibsp','parch','family']].head()
```

```python
def get_category(x):
    cat = ''
    if x <= 10:
        cat='Child'
    elif x <= 30:
        cat='Young Adult'
    elif x <= 60:
        cat ='Adult'
    else: 'Elderly'
    return cat

df['age_class'] = df.age.apply(lambda x:get_category(x))
df.head()
```
