```python
# 배열 생성 및 기본 작업
import numpy as np

# Creating a NumPy array
array = np.array([1, 2, 3, 4])

# Basic operations
print("Original array:", array)
print("Array + 2:", array + 2)
print("Array squared:", array ** 2)

# Basic statistics
print("Mean of array:", np.mean(array))
```

```python
# 배열 생성 및 속성 탐색:
# Create a 1D array from a list
array_1d = np.array([3, 6, 9, 12])
print("1D Array:", array_1d)

# Array attributes
print("Shape:", array_1d.shape) # 배열의 크기
print("Size:", array_1d.size) # 배열의 총 요소 수
print("Type:", array_1d.dtype) # 배열 요소의 데이터 유형

# Create a 2D array using a nested list
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:\n", array_2d)
print("Shape:", array_2d.shape)
```

```python
# 인덱싱 및 슬라이싱:
# Indexing
print("\nElement at index 2:", array_1d[2])

# Slicing
print("Elements from index 1 to 3:", array_1d[1:4])

# 2D Array Indexing
print("\nElement in 2nd row, 3rd column:", array_2d[1, 2])

# Slicing in 2D
print("First row:", array_2d[0, :])
```

```python
# Arithmetic Operations
array = np.array([1, 2, 3, 4])
print("\nArray:", array)
print("Array + 2:", array + 2)
print("Array squared:", array ** 2)

# Universal functions (ufuncs)
print("Square root:", np.sqrt(array))
```

선형대수학은 과학과 공학의 여러 분야에 기초가 되는 수학의 한 분야. 벡터, 행렬, 선형 변환을 다룬다.

1. 벡터와 공간
벡터: 크기와 방향을 모두 갖는 객체입니다. n차원 공간으로 표현.
벡터 공간: 동일한 공간에서 다른 벡터를 생성하기 위해 함께 더하고 스칼라를 곱할 수 있는 벡터 집합.
2. 행렬
행렬: 행과 열로 배열된 직사각형 숫자 배열. 선형 변환을 나타낸다.
행렬 연산: 덧셈, 뺄셈, 곱셈, 나눗셈(역전)을 포함.
전치: 행렬을 대각선 위로 뒤집는다.
3. 선형 방정식 시스템
선형 방정식 모음. 행렬을 사용하여 표현하고 풀 수 있다.
4. 행렬식
정사각 행렬 항목의 함수인 스칼라 값. 행렬 가역성을 이해하는 데 중요.
5. 고유값과 고유벡터
고유값: 선형 방정식 시스템과 관련된 스칼라.
고유벡터: 선형 변환이 적용될 때 최대 스칼라 인수만큼 변경되는 0이 아닌 벡터.
6. 행렬 분해
행렬을 행렬의 곱으로 분해. 예로는 LU 분해, QR 분해, SVD(특이값 분해)가 있다.
7. 응용 프로그램
3D 모델링, 컴퓨터 그래픽, 엔지니어링, 물리학, 기계 학습 등과 같은 분야에서 널리 사용.

```python
#  NumPy의 선형 대수학 모듈을 사용하여 간단한 선형 방정식 시스템을 푸는 방법

import numpy as np

# Coefficient matrix and constant array
A = np.array([[3, 2], [1, -1]])
B = np.array([18, 1])

# Solve the system of equations
solution = np.linalg.solve(A, B)
print("Solution:", solution)
```

```python
# 선형대수학

# Matrix multiplication
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# Dot product
print("Dot Product:\n", np.dot(a, b))

# Eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(a)
print("Eigenvalues:", eigenvalues)

# Matrix Inversion
inverse = np.linalg.inv(a)
print("Inverse of a:\n", inverse)
```

뷰(view) : 넘파이의 ndarray(이하 배열)을 슬라이싱할 때 파이썬의 리스트(list)와 다르게 원본의 참조가 생성

numpy는 대용량 데이터 처리를 염두에 두고 대용량 데이터의 슬라이싱이 빈번하게 일어나는 복잡한 코드를 실행시키는데 있어서 복사가 남발되면 메모리 문제를 일으킬 소지가 많기 때문에 이렇게 설계

- 복사가 아니고 뷰(참조의 개념)이기 때문에 원본이 바뀌면 이 참조본도 따라서 바뀜
- 리스트의 슬라이싱에서는 복사본이 생성되는 것과는 다른 동작이기 때문에 주의
- 슬라이싱의 복사본을 생성하고 싶다면 copy() 속성을 이용
