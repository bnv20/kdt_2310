#### re 모듈

- 파이썬에서 정규 표현식(Regular Expressions)을 지원하기 위한 표준 라이브러리.
- 정규 표현식은 텍스트에서 문자열 검색, 매칭, 치환 및 파싱을 할 때 사용하는 강력한 문자열 처리 도구

re 모듈을 사용하여 수행할 수 있는 기본 작업

- re.match(): 문자열의 시작부터 정규식과 일치하는 패턴을 찾습니다. 일치하는 객체를 반환하거나, 일치하는 것이 없으면 None을 반환

- re.search(): 문자열 전체에 걸쳐 정규식과 일치하는 패턴을 찾는다. match와 달리 문자열의 중간부터 일치하는 패턴을 찾을 수도 있고 일치하는 객체를 반환하거나, 일치하는 것이 없으면 None을 반환

- re.findall(): 문자열 내에서 정규식과 일치하는 모든 부분 문자열을 리스트로 반환

- re.finditer(): findall()과 유사하지만, 일치하는 모든 결과를 이터레이터(iterator)로 반환하여 메모리를 절약

- re.sub(): 문자열 내에서 정규식과 일치하는 부분을 다른 문자열로 치환

- re.compile(): 정규식 패턴을 컴파일하여 나중에 재사용할 수 있는 정규식 객체를 생성. 이것은 패턴을 여러 번 사용할 때 성능을 최적화하는 데 도움

정규 표현식 패턴은 특수 문자와 시퀀스를 포함할 수 있으며, 복잡한 문자열 처리를 위한 강력한 문법을 제공.
예를 들어, \d는 숫자를, \w는 단어 문자를, .는 거의 모든 문자를 나타낸다(개행 문자는 제외). 대괄호([])는 문자 집합을, 
괄호(())는 그룹을 정의한다. 별표(*), 더하기(+), 물음표(?), 중괄호({}) 등은 일치하는 횟수를 정의.

```python
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']

#양 옆 스페이스 먼저 지우기
n_states = [state.strip() for state in states]

result_states = []

for state in n_states:
    if state.isalpha()==False:
        space_index = state.find(' ')
        temp_state = list(state)
        new_state = ""
        for i in range(len(temp_state)):
            if temp_state[i].isalpha():
                new_state += temp_state[i]
            if i == space_index:
                new_state += temp_state[i]
        result_states.append(new_state.title())
    else:
        result_states.append(state.title())

#리스트 형식에 맞게 세로로 출력
separate = '\',\n\''
separate_result = separate.join(result_states)
print(f'[\'{separate_result}\']')
```

#### Python random 모듈 
Python의 표준 라이브러리 중 하나로, 다양한 유형의 무작위 데이터를 생성하기 위한 함수와 도구들을 제공합니다. 여기에는 단일 숫자, 숫자 범위, 무작위 순서로 요소를 선택하거나 배열을 섞는 기능 등이 포함
1. random.random() : 0.0에서부터 1.0 사이의 실수(float)를 반환. 0은 포함되고 1은 포함되지 않음
2. random.uniform(a, b) : 인자로 들어온 a~b 사이의 실수(float)를 반환
3. randint(a, b) : 인자로 들어온 a, b 사이의 랜덤한 정수(int)를 반환
4. randrange(a, b), randrange(b) :  a <= x < b 의 범위 내에서의 랜덤한 정수(int)를 반환, randrange(b)는 0 <= x < b 의 범위 내에서의 랜덤한 정수(int)를 반환
5. random.choice(seq) : 인자로 받은 리스트, 튜플, 문자열, range 에서 무작위로 하나의 원소 반환
6. random.sample(seq or set, N) : 첫번째 인자로 받은 시퀀스데이터 or set에서 N개의 랜덤하고, unique하고, 순서상관없이 인자를 뽑아서 리스트로 만들어서 반환. 예: random.sample([1, 2, 3, 4, 5], 3)은 5개의 숫자 중 3개를 임의로 선택한 리스트를 반환
7. random.shuffle(seq) : 데이터의 순서를 무작위로 랜덤하게 바꾸어 주는 함수

```python
import random

# 인풋은 추가할 메뉴(str), 메뉴리스트(set)
def add_menu_items(item,menu):
    menu.add(item)
    return menu

# 인풋은 메뉴리스트(set), 추천수(숫자)
def get_recommendation(menu,recommend_num):
    rec_menu = random.sample(sorted(menu),recommend_num)
    return rec_menu

    
def main():
    menu = {'맥도날드', '쉑쉑버거', '편의점', '분식집', '닭곰탕', '짬뽕'}
    # 메뉴 추가
    while 1:
        menu_input = input(f"주변 음식점 이름을 적으세요. 없으면 'stop'을 입력하세요: ")
        if menu_input == 'stop':
            break
        else:
            menu = add_menu_items(menu_input,menu)
            print(f"메뉴가 성공적으로 추가되었습니다.\n<메뉴목록>\n{menu}")

    # 메뉴 추천
    end_chr = 1 # 1==메뉴 추천 루프를 실행, 0==메뉴 추천 루프를 종료
    while end_chr == 1:
        menu_num = 0
        while 1:
            menu_num_str = input("추천받고 싶은 메뉴의 갯수를 말하세요 (숫자만): ")
            if menu_num_str.isdigit() == False:
                print("잘못 입력하셨습니다. 숫자만 입력해주세요")
                continue
            if int(menu_num_str) > len(menu):
                print("잘못 입력하셨습니다. {menu.len()} 이하의 숫자만 입력해주세요")
                continue
            menu_num = int(menu_num_str)
            break
        while 1:
            rec_menu = get_recommendation(menu,menu_num)
            print(f'추천 메뉴는 {rec_menu} 입니다.')
            diff_menu_yn = input("\n다른 추천을 원하시나요? (y/n) ")
            if diff_menu_yn == "y":
                for i in rec_menu:
                    menu.remove(i)   # 중복 추천을 피할 수 있게 추천하는 메뉴 삭제
                if len(menu) ==0:
                    print("더이상 추천할 수 있는 메뉴가 없습니다.")
                    end_chr = 0
                    break
                menu_num = random.randint(1,len(menu))
                print(f"추천 메뉴의 수를 {menu_num}으로 설정합니다.")
            elif diff_menu_yn == "n":
                print("메뉴 추천을 종료합니다.")
                end_chr = 0
                break
            else:
                print("죄송합니다. 잘못입력하셨으므로 메뉴 추천을 종료합니다.")
                end_chr = 0
                break
                
main()
```

```python
# 사용자 함수 활용

import random

def add_menu_items(menu):
  print(f'기본으로 제공하는 메뉴:\n{menu}\n')
  while True:
    user_input = input("주변 음식점 이름을 적으세요. 없으면 'stop'을 입력: ").strip()
    if user_input.lower() == 'stop':
      break
    else:
      menu.add(user_input)
  return list(menu)
def get_recommendation(menu, num_items):
  try:
    picks = random.sample(menu, num_items)
    print(f'추천 메뉴는 {picks} 입니다.\n')
  except ValueError:
    print(f'숫자가 너무 큽니다. 메뉴는 {len(menu)}개밖에 없습니다.')

def main():
  menu = {"짬뽕", "분식집", "맥도날드", "쉑쉑버거", "편의점", "닭곰탕"}
  menu = add_menu_items(menu)
  while True:
    try:
      num_recommendations = int(input('추천받고 싶은 메뉴의 개수를 입력하세요: '))
      get_recommendation(menu,num_recommendations)
      break
    except ValueError:
      print(f'숫자를 적어주세요')
  choice = input('다른 추천을 원하시나요? (y/n): ').strip().lower()
  if choice == 'y':
    recmd_num = random.randrange(1,len(menu)+1)   
    print(f'제가 추천하는 메뉴수는 다음과 같이 {recmd_num}개 입니다')
    get_recommendation(menu,recmd_num)
  elif choice == 'n':
    print('식사 맛있게 하세요~')
  else:
    print('잘못된 입력입니다. 프로그램 종료합니다.')

if __name__ == "__main__":
  main()
```

```python
import math
def area_circumference(r):
    area = math.pi*r*r
    circumference = 2*math.pi*r
    print(f"면적: {round(area,2)}, 둘레: {round(circumference,2)}")

area_circumference(5)
area_circumference(10)
```

파이썬 if __name__ == "__main__" 의미

- 메인 함수의 선언, 시작을 의미합니다. 해당 코드 밑에 main 등의 함수 호출 코드를 작성해서 함수의 기능을 수행
- 직접 modex2.py 파일을 실행하는 경우에는 __name___ 변수에 __main__ 이라는 값이 할당되고 이 조건절이 참이 되어 if문 다음 문장들이 수행
- 모듈을 불러와서 사용하는 경우 __name__ 변수에는 모듈 이름이 저장되고 이 조건절이 거짓이 되어 if문 다음 문장들이 수행되지 않음
- if __name__ == "__main__"와 같이 조건문을 사용하여 터미널에서와 같이 직접 호출되어 사용될 때는 그 자체로 기능을 수행하고, 동시에 다른 모듈에서 필요한 함수 등을 제공

```python
# my_module.py

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

# 아래 if 블록은 my_module.py가 직접 실행될 때만 실행됩니다.
if __name__ == "__main__":
    print("Module is being run directly")
    print("Add 3 + 5 =", add(3, 5))
    print("Multiply 4 * 7 =", multiply(4, 7))
```

#### 날짜 함수

날짜와 시간을 처리하기 위한 주요 라이브러리는 datetime입니다. 이 라이브러리는 날짜, 시간, 시간대와 관련된 다양한 클래스와 함수를 제공하며, 날짜와 시간 연산에 매우 유용
- datetime.date: 날짜(연, 월, 일)를 표현하는 클래스
    - today(): 오늘의 날짜를 반환
- datetime.time: 시간(시, 분, 초, 마이크로초)을 표현하는 클래스
- datetime.datetime: 날짜와 시간을 모두 포함하는 클래스
    - now(): 현재의 날짜와 시간을 반환
- datetime.timedelta: 두 날짜/시간 간의 차이를 표현하는 클래스
- datetime.tzinfo: 시간대 정보를 위한 기본 클래스

datetime 모듈의 날짜와 시간 객체들을 문자열 형식으로 포맷팅하거나 문자열로부터 날짜와 시간 객체를 생성하기 위해 사용하는 주요 메서드는 strftime()와 strptime()
- strftime() (String Format Time)
    - 날짜와 시간 객체를 문자열로 변환하는 데 사용
    - 사용자가 지정한 형식 문자열을 기반으로 날짜와 시간 정보를 문자열로 표현
- strptime() (String Parse Time)
    - 문자열을 날짜와 시간 객체로 변환하는 데 사용
    - 주어진 형식 문자열을 기반으로 문자열에서 날짜와 시간 정보를 파싱 

```python
import datetime

# 현재 날짜와 시간 가져오기
now = datetime.datetime.now()
print(now)  # 예: 2023-11-02 14:29:45.678123

# 오늘의 날짜만 가져오기
today = datetime.date.today()
print(today)  # 예: 2023-11-02

# 특정 날짜와 시간을 직접 지정
d = datetime.datetime(2023, 11, 2, 14, 29, 45)
print(d)  # 출력: 2023-11-02 14:29:45

# 날짜 간의 차이 계산
delta = datetime.timedelta(days=7)  # 7일의 기간
new_date = today + delta
print(new_date)  # 오늘로부터 7일 후의 날짜

# 두 날짜 간의 차이 계산
date1 = datetime.date(2023, 1, 1)
date2 = datetime.date(2023, 12, 31)
diff = date2 - date1
print(diff.days)  # 출력: 364
```

과제9_1103 : 파이썬의 random 모듈을 사용하여 간단한 구매 데이터 세트를 생성하세요. 이 데이터 세트에는 다음과 같은 정보가 포함:

- 구매 ID
- 구매자 ID
- 구매한 제품의 ID
- 구매한 제품의 수량
- 구매 날짜와 시간
- purchase_id: 1000에서 9999 사이의 랜덤 정수
- buyer_id: 100에서 999 사이의 랜덤 정수
- product_id: 10에서 99 사이의 랜덤 정수
- quantity: 1에서 10 사이의 랜덤 정수
- purchase_date: 최근 365일 이내의 랜덤 날짜 및 시간

```python
import random
from datetime import datetime, timedelta

# 가상의 데이터를 생성하는 함수
def generate_purchase_data(num_records):
    purchase_data = []

    for _ in range(num_records):
        purchase_id = random.randint(1000, 9999)
        buyer_id = random.randint(100, 999)
        product_id = random.randint(1000, 9999)
        quantity = random.randint(1, 10)
        
        # 현재 시간으로부터 최대 365일 이전의 날짜를 무작위로 생성
        days_ago = random.randint(0, 365)
        purchase_date = datetime.now() - timedelta(days=days_ago)

        purchase_record = {
            'purchase_id': purchase_id,
            'buyer_id': buyer_id,
            'product_id': product_id,
            'quantity': quantity,
            'purchase_date': purchase_date.strftime('%Y-%m-%d %H:%M:%S')  # 포맷팅된 문자열
        }

        purchase_data.append(purchase_record)
    
    return purchase_data

# 데이터 세트 생성 예제: 10개의 레코드를 생성
purchase_dataset = generate_purchase_data(10)

# 생성된 데이터 세트 출력
for record in purchase_dataset:
    print(record)
```

#### 파이썬에서 언더바/더블언더바 의 의미와 역할

단일 언더바 (_):

- 인터프리터: 파이썬 인터프리터에서 단일 언더바는 가장 최근의 표현식의 결과를 저장
- 변수 이름: 관례적으로 관심이 없는 값을 할당할 때 사용합니다. 예를 들어, for _ in range(5):는 반복문에서 인덱스를 사용하지 않을 때 관용적으로 사용
- 국제화(i18n)/지역화(l10n): 함수 이름으로서 _()는 종종 gettext 모듈을 통해 문자열을 번역하는 데 사용
- 값 무시: 튜플 언패킹에서 특정 값을 무시할 때 사용합니다. 예: x, _, y = (1, 2, 3) 여기서 2는 무시

단일 선행 언더바 (_변수):
- 일반적으로 모듈 내부에서만 사용되는 "private" 변수나 함수 또는 메서드를 지정할 때 사용
- 파이썬에서 강제할 수는 없지만, 이것은 프로그래머에게 그 변수나 함수가 내부 사용을 위한 것임을 알려주는 관례

단일 후행 언더바 (변수_):
- 파이썬 키워드와 충돌을 피하기 위해 사용됩니다. 예를 들어, class_는 class라는 이름 대신 사용

더블 선행 언더바 (__변수):
- 클래스의 속성을 이름 맹글링(name mangling)하기 위해 사용
- 이는 속성이 해당 클래스와 그 서브클래스 내부에서만 사용될 것임을 나타내며, 부모 클래스의 필드와 이름 충돌을 피하는 데 도움을 준다. 예: __private_var는 _ClassName__private_var로 변환되어 클래스 외부에서 직접 접근할 수 없습니다.

더블 선행과 후행 언더바 (변수):
- 파이썬의 매직 메서드 또는 특별 메서드를 나타냅니다. 이들은 특별한 의미를 가지며, 파이썬의 내부 구조와 연동
- 예를 들어, __init__, __call__, __str__, __repr__ 등이 있습니다. 이러한 메서드들은 사용자 정의 객체가 내장 타입과 상호작용하는 방식을 정의<br>
    __init__: 객체가 생성될 때 초기화를 담당합니다.<br>
    __repr__: 객체의 “공식적인” 문자열 표현을 제공합니다.<br>
    __add__: + 연산자로 두 객체를 더할 때 호출됩니다.<br>
    __call__: 객체를 함수처럼 호출할 수 있게 합니다 (예: instance()).<br>
    __getitem__: 인덱싱을 통해 요소에 접근할 때 호출됩니다 (예: instance[key]).
- __main__은 파이썬에서 사용되는 특별한 변수이며, 현재 최상위 실행 환경이 무엇인지를 나타내는 데 사용됩니다. 파이썬 스크립트가 직접 실행될 때 그 스크립트의 __name__ 변수는 "__main__"으로 설정

#### 과제 
카운터에서 거스름돈으로 사용할 500원, 100원, 50원, 10원짜리 동전이 무한히 존재한다고 가정한다. 손님에게 거슬러 줘야 할 돈이 N원일 때 거슬러줘야 할 동전의 최소 개수를 구하세요. 거슬러 줘야 할 돈 N은 항상 10의 배수이다. N은 2750원이다.

```python
n = 2750
count = 0

# 큰 단위의 화폐부터 차례대로 확인하기
coin_types = [500, 100, 50, 10]

for coin in coin_types:
    count += n // coin # 해당 화폐로 거슬러 줄 수 있는 동전의 개수 세기
    n %= coin

print(count)
```

#### 과제 
파일이름과 특정 단어를 인자로 입력하면 그 파일에 있는 특정 단어의 개수를 반환하는 사용자 함수를 이용하여 mydata 파일에서 how의 개수를 출력하세요.  

```python
def countWord(filename, word):
    with open(filename, 'r') as f:
        text = f.read()
        text = text.lower()
        pos = text.find(word) # 인덱스
        count = 0
        while pos != -1:
            count += 1
            pos = text.find(word, pos+1)  # string.find(찾을 문자, 시작 Index, 끝 Index)
    return count

filename = input('filename을 입력하세요: ')
word = input('filename에서 개수를 구할 단어를 입력하세요: ')
ret = countWord(filename, word)
print(f'{word}의 개수: {ret}')
```

#### 파이썬에서 -1은 다양한 방식으로 사용됩니다. 다음은 일부 예시입니다:

- 인덱스: 파이썬에서는 리스트, 문자열, 튜플 등의 시퀀스 자료형의 인덱스를 지정할 때 음수를 사용할 수 있습니다. 이때 -1은 시퀀스의 마지막 요소를 가리킵니다. 예를 들어, 리스트에서 my_list[-1]은 리스트의 마지막 요소를 반환합니다.

- 슬라이스: 마찬가지로, 시퀀스 자료형의 슬라이스를 지정할 때도 음수를 사용할 수 있습니다. 이때 -1은 시퀀스의 마지막 요소를 가리킵니다. 예를 들어, 리스트에서 my_list[:-1]은 마지막 요소를 제외한 모든 요소를 반환합니다.

- 반환 값: 일부 함수에서는 실패 또는 오류를 나타내는 값을 -1로 반환합니다. 예를 들어, 문자열에서 find() 메서드는 문자열에서 지정된 문자열을 찾을 수 없을 때 -1을 반환합니다.

- 비트 연산: 비트 연산에서는 -1이 모든 비트가 1인 이진수로 사용됩니다. 예를 들어, 0b11111111은 10진수로 255이며, -1과 같습니다.
