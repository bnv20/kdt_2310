## Python

- Draw Object tracking trails

```python
import cv2
import numpy as np
from ultralytics import YOLO

from ultralytics.utils.checks import check_imshow
from ultralytics.utils.plotting import Annotator, colors # AnnotatorëŠ” ì´ë¯¸ì§€ì— ëŒ€í•œ ì£¼ì„(ì˜ˆ: ë°”ìš´ë”© ë°•ìŠ¤, í…ìŠ¤íŠ¸ ë¼ë²¨)ì„ ì¶”ê°€í•˜ëŠ” ë° ì‚¬ìš©ë˜ë©°, colors í•¨ìˆ˜ëŠ” ê°ì²´ íƒì§€ì—ì„œ ì‚¬ìš©í•  ìƒ‰ìƒì„ ì œê³µ
from collections import defaultdict # ê°ì²´ ì¶”ì ì—ì„œ ê° ê°ì²´ì˜ ì´ë™ ê²½ë¡œë¥¼ ì €ì¥

track_history = defaultdict(lambda: []) # ìƒˆ í‚¤ì— ëŒ€í•œ ê¸°ë³¸ê°’ìœ¼ë¡œ ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ìƒì„±í•˜ëŠ” ëŒë‹¤ í•¨ìˆ˜
model = YOLO("yolov8n.pt")
names = model.model.names # YOLO ëª¨ë¸ ë‚´ì—ì„œ íƒì§€í•  ìˆ˜ ìˆëŠ” ê°ì²´ í´ë˜ìŠ¤ì˜ ì´ë¦„ì„ í¬í•¨í•˜ëŠ” ë¦¬ìŠ¤íŠ¸ë¥¼ names ë³€ìˆ˜ì— í• ë‹¹
video_path = "/content/downloaded_video.mp4"
cap = cv2.VideoCapture(video_path) # ì´ ê°ì²´ëŠ” ë¹„ë””ì˜¤ ìŠ¤íŠ¸ë¦¼ì„ ì œì–´í•˜ê³ , í”„ë ˆì„ë³„ë¡œ ì½ê¸° ì‘ì—…ì„ ìˆ˜í–‰í•˜ëŠ” ë° ì‚¬ìš©
assert cap.isOpened(), "Error reading video file" # ë¹„ë””ì˜¤ íŒŒì¼ì´ ì„±ê³µì ìœ¼ë¡œ ì—´ë ¸ëŠ”ì§€ í™•ì¸
# cap.get() ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ë¹„ë””ì˜¤ì˜ ë„ˆë¹„(cv2.CAP_PROP_FRAME_WIDTH), ë†’ì´(cv2.CAP_PROP_FRAME_HEIGHT), ê·¸ë¦¬ê³  í”„ë ˆì„ ë ˆì´íŠ¸(cv2.CAP_PROP_FPS)ë¥¼ ì¶”ì¶œ
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
# ì²˜ë¦¬ëœ ë¹„ë””ì˜¤ í”„ë ˆì„ì„ ìƒˆë¡œìš´ íŒŒì¼ë¡œ ì €ì¥
result = cv2.VideoWriter("object_tracking.avi",
                       cv2.VideoWriter_fourcc(*'mp4v'), # ë¹„ë””ì˜¤ ì½”ë”© í˜•ì‹ì„ ì§€ì •. 'mp4v' ì½”ë±ì„ ì‚¬ìš©
                       fps,
                       (w, h))

while cap.isOpened():
    success, frame = cap.read()
    if success:
        results = model.track(frame, persist=True, verbose=False)
        boxes = results[0].boxes.xyxy.cpu() # í˜„ì¬ ë¹„ë””ì˜¤ í”„ë ˆì„ ë‚´ì—ì„œ íƒì§€ ë° ì¶”ì ëœ ëª¨ë“  ê°ì²´ë“¤ì˜ ë°”ìš´ë”© ë°•ìŠ¤ ì¢Œí‘œë¥¼ xyxy í˜•ì‹ìœ¼ë¡œ CPU ë©”ëª¨ë¦¬ì— ì €ì¥í•˜ëŠ” ì‘ì—…ì„ ìˆ˜í–‰

        if results[0].boxes.id is not None:

            # íƒì§€ëœ ê° ê°ì²´ì— ëŒ€í•´, ë°”ìš´ë”© ë°•ìŠ¤(boxes), í´ë˜ìŠ¤ ID(class), ì¶”ì  ID(track_ids), ì‹ ë¢°ë„ ì ìˆ˜(confs)ë¥¼ ì¶”ì¶œ
            clss = results[0].boxes.cls.cpu().tolist()
            track_ids = results[0].boxes.id.int().cpu().tolist()
            confs = results[0].boxes.conf.float().cpu().tolist()

            # Annotator í´ë˜ìŠ¤ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¹„ë””ì˜¤ í”„ë ˆì„ì— ì‹œê°ì  ì£¼ì„(ì˜ˆ: ë°”ìš´ë”© ë°•ìŠ¤, í…ìŠ¤íŠ¸ ë¼ë²¨)ì„ ì¶”ê°€í•˜ê¸° ìœ„í•œ ê°ì²´ë¥¼ ìƒì„±
            annotator = Annotator(frame, line_width=2)

            # box_label ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ê° ê°ì²´ì— ëŒ€í•œ ë°”ìš´ë”© ë°•ìŠ¤ì™€ ë¼ë²¨ì„ í”„ë ˆì„ì— ì¶”ê°€
            for box, cls, track_id in zip(boxes, clss, track_ids):
                annotator.box_label(box, color=colors(int(cls), True), label=names[int(cls)])

                # ê° ê°ì²´ì— ëŒ€í•´ ê·¸ ê°ì²´ì˜ ì´ë™ ê²½ë¡œë¥¼ ì‹œê°ì ìœ¼ë¡œ í‘œí˜„
                track = track_history[track_id]
                # ê°ì²´ì˜ í˜„ì¬ ìœ„ì¹˜(ë°”ìš´ë”© ë°•ìŠ¤ì˜ ì¤‘ì‹¬ ì¢Œí‘œ)ë¥¼ ì¶”ì  ê²½ë¡œì— ì¶”ê°€. ë°”ìš´ë”© ë°•ìŠ¤ì˜ ì¤‘ì‹¬ ì¢Œí‘œëŠ” ì¢Œìƒë‹¨(box[0], box[1])ê³¼ ìš°í•˜ë‹¨(box[2], box[3]) ì¢Œí‘œì˜ í‰ê· ìœ¼ë¡œ ê³„ì‚°
                track.append((int((box[0] + box[2]) / 2), int((box[1] + box[3]) / 2)))
                if len(track) > 30: # ì¶”ì  ê²½ë¡œì˜ ê¸¸ì´ê°€ 30ì„ ì´ˆê³¼í•˜ëŠ” ê²½ìš°, ê°€ì¥ ì˜¤ë˜ëœ ìœ„ì¹˜ ì •ë³´ë¥¼ ì œê±°
                    track.pop(0)

                # Plot tracks
                points = np.array(track, dtype=np.int32).reshape((-1, 1, 2))
                # í˜„ì¬ í”„ë ˆì„ì—ì„œ ê°ì²´ì˜ ìµœì‹  ìœ„ì¹˜ì— ì›ì„ ê·¸ë¦½ë‹ˆë‹¤. ì›ì˜ ì¤‘ì‹¬ì€ ì¶”ì  ê²½ë¡œì˜ ë§ˆì§€ë§‰ ì¢Œí‘œ(track[-1])ì´ë©°, ìƒ‰ìƒì€ ê°ì²´ì˜ í´ë˜ìŠ¤ IDì— ë”°ë¼ ê²°ì •
                cv2.circle(frame, (track[-1]), 7, colors(int(cls), True), -1)
                cv2.polylines(frame, [points], isClosed=False, color=colors(int(cls), True), thickness=2) # ì¶”ì  ê²½ë¡œë¥¼ ì„ ìœ¼ë¡œ ê·¸ë ¤ ê°ì²´ì˜ ì´ë™ ê²½ë¡œë¥¼ ì‹œê°í™”

        result.write(frame)
        if cv2.waitKey(1) & 0xFF == ord("q"): # ì‚¬ìš©ìê°€ "q" í‚¤ë¥¼ ëˆŒëŸ¬ ì–¸ì œë“ ì§€ ì¢…ë£Œ
            break
    else:
        break

result.release()
cap.release()
cv2.destroyAllWindows()
```
