- Ultralytics yolo 명령은 다음 구문을 사용
  - TASK (선택 사항)은 다음 중 하나입니다. [detect, segment, classify]. 명시적으로 전달되지 않은 경우 YOLOv8 은 문자열의 TASK 를 모델 유형에서 선택합니다.
  - MODE (필수) 중 하나입니다. [train, val, predict, export, track]
  - ARGS (선택 사항)은 사용자 정의 arg=value 같은 쌍 imgsz=320 를 사용하여 기본값을 재정의할 수 있습니다. 사용 가능한 전체 목록은 ARGS 를 참조하십시오

## Python

- Draw Object tracking trails

```python
import cv2
import numpy as np
from ultralytics import YOLO

from ultralytics.utils.checks import check_imshow
from ultralytics.utils.plotting import Annotator, colors # Annotator는 이미지에 대한 주석(예: 바운딩 박스, 텍스트 라벨)을 추가하는 데 사용되며, colors 함수는 객체 탐지에서 사용할 색상을 제공
from collections import defaultdict # 객체 추적에서 각 객체의 이동 경로를 저장

track_history = defaultdict(lambda: []) # 새 키에 대한 기본값으로 빈 리스트를 생성하는 람다 함수
model = YOLO("yolov8n.pt")
names = model.model.names # YOLO 모델 내에서 탐지할 수 있는 객체 클래스의 이름을 포함하는 리스트를 names 변수에 할당
video_path = "/content/downloaded_video.mp4"
cap = cv2.VideoCapture(video_path) # 이 객체는 비디오 스트림을 제어하고, 프레임별로 읽기 작업을 수행하는 데 사용
assert cap.isOpened(), "Error reading video file" # 비디오 파일이 성공적으로 열렸는지 확인
# cap.get() 메서드를 사용하여 비디오의 너비(cv2.CAP_PROP_FRAME_WIDTH), 높이(cv2.CAP_PROP_FRAME_HEIGHT), 그리고 프레임 레이트(cv2.CAP_PROP_FPS)를 추출
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
# 처리된 비디오 프레임을 새로운 파일로 저장
result = cv2.VideoWriter("object_tracking.avi",
                       cv2.VideoWriter_fourcc(*'mp4v'), # 비디오 코딩 형식을 지정. 'mp4v' 코덱을 사용
                       fps,
                       (w, h))

while cap.isOpened():
    success, frame = cap.read()
    if success:
        results = model.track(frame, persist=True, verbose=False)
        boxes = results[0].boxes.xyxy.cpu() # 현재 비디오 프레임 내에서 탐지 및 추적된 모든 객체들의 바운딩 박스 좌표를 xyxy 형식으로 CPU 메모리에 저장하는 작업을 수행

        if results[0].boxes.id is not None:

            # 탐지된 각 객체에 대해, 바운딩 박스(boxes), 클래스 ID(class), 추적 ID(track_ids), 신뢰도 점수(confs)를 추출
            clss = results[0].boxes.cls.cpu().tolist()
            track_ids = results[0].boxes.id.int().cpu().tolist()
            confs = results[0].boxes.conf.float().cpu().tolist()

            # Annotator 클래스를 사용하여 비디오 프레임에 시각적 주석(예: 바운딩 박스, 텍스트 라벨)을 추가하기 위한 객체를 생성
            annotator = Annotator(frame, line_width=2)

            # box_label 메서드를 사용하여 각 객체에 대한 바운딩 박스와 라벨을 프레임에 추가
            for box, cls, track_id in zip(boxes, clss, track_ids):
                annotator.box_label(box, color=colors(int(cls), True), label=names[int(cls)])

                # 각 객체에 대해 그 객체의 이동 경로를 시각적으로 표현
                track = track_history[track_id]
                # 객체의 현재 위치(바운딩 박스의 중심 좌표)를 추적 경로에 추가. 바운딩 박스의 중심 좌표는 좌상단(box[0], box[1])과 우하단(box[2], box[3]) 좌표의 평균으로 계산
                track.append((int((box[0] + box[2]) / 2), int((box[1] + box[3]) / 2)))
                if len(track) > 30: # 추적 경로의 길이가 30을 초과하는 경우, 가장 오래된 위치 정보를 제거
                    track.pop(0)

                # Plot tracks
                points = np.array(track, dtype=np.int32).reshape((-1, 1, 2))
                # 현재 프레임에서 객체의 최신 위치에 원을 그립니다. 원의 중심은 추적 경로의 마지막 좌표(track[-1])이며, 색상은 객체의 클래스 ID에 따라 결정
                cv2.circle(frame, (track[-1]), 7, colors(int(cls), True), -1)
                cv2.polylines(frame, [points], isClosed=False, color=colors(int(cls), True), thickness=2) # 추적 경로를 선으로 그려 객체의 이동 경로를 시각화

        result.write(frame)
        if cv2.waitKey(1) & 0xFF == ord("q"): # 사용자가 "q" 키를 눌러 언제든지 종료
            break
    else:
        break

result.release()
cap.release()
cv2.destroyAllWindows()
```
